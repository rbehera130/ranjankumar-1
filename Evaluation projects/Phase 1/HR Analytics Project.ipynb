### Conclusion

By visualizing the data we have reached to the following conclusions:
    
    1: There are various reasons which are a factor of employee attrition rate.
    2: These can't be completely avoided but companies can work on them and it may reduce the attrition rate.
    3: Employess switch to the other companies for better salary hike.
    4: They switch to for good worklife balance although the worklife balance in their current company is quite good but
       instead of that they tend to switch so they can get more better worklife balance.
    5: Sometimes employees don't satisfy with their work environment and that could be the reason to leave.

### Some Strategies that may reduce attrition rate:

By analysing the data we have seen that it may not be possible to avoid some points which are cause of attrition of the employees like employees with technical degree have a high attrition rate as compare to employee with other education background. But a company can not avoid them they can't say that we will not hire technical background people because to do their work they need them.

So, there are certain other things or strategies that companies can apply and that may reduce the attrition rate of employees these are following:

    1: Companies can provide an acceptable salary hike to their current employees who have good performance rating because
       we have seen employees who get a better salary there is less chance of their attrition as they switch for a better 
       salary hike.
    2: Companies can try and create a good working environment so that they can reduce the chance of attrition and it will
       also play a big role in the growth of a company.
    3: Every employee wants a better worklife balance although they are getting it but for the want of more better they
       tend to switch to the other company so companies can work on the direction to provide a very well worklife balance
       to their employees.
