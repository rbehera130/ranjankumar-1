## Bivariate Analysis

### Let's check the diseases which cause most and least number of people death

```python
# As our dataset is big so making empty lists for death count and disease, using for loop to find and adding total number of 
# deaths due to every disease then making a dictionary using them so that we can make a dataframe of that and plot it on graph
# for analysis purpose

disease = columns[3:]  # adding disease names in list
total_deaths = []
for i in columns[3:]:
    total_deaths.append(data[i].sum())  # finding total death count and adding it into list
    
# making dictionary of disease and death counts
total_figure = {'Disease': disease, 'Total_deaths': total_deaths}
```

```python
# making dataframe of our dictionary
death_counts = pd.DataFrame(total_figure)

# checking top 5 rows of our dataframe
death_counts.head()
```

```python
# our data is ready now we can plot it into graph but before that we will sort our data in descending number of death counts
# so that we will get the disease due to which highest and lowest number of people are dead
sorted_frame = death_counts.sort_values(by = 'Total_deaths', ascending = False)

# now or dataframe is sorted in descending number of death counts
```

```python
# finding top 10 disease due to which most number of people dies
plt.figure(figsize = (15,10))
plt.title('Top death cause')
sns.barplot(x = "Total_deaths", y = 'Disease', data = sorted_frame.iloc[ :10, :], orient = "h")
plt.show()
```

These are the top 10 disease due to which most number of people died in all countries.

```python
# finding top 10 disease which are the least factors of people death
plt.figure(figsize = (15,10))
plt.title('Least death cause')
sns.barplot(x = "Total_deaths", y = 'Disease', data = sorted_frame.iloc[-10:, :], orient = "h")
plt.show()
```

These are least 10 factors of people death.

### Let's check countries with high and low number of deaths

```python
# first let's group countries because these contains data year by year
countries_deaths = data.groupby('Country/Territory').sum()
```

```python
# we don't need year column so let's drop it
countries_deaths.drop('Year', axis = 1, inplace = True)
```

```python
# adding countries name in a list
countries = list(countries_deaths.index)
```

```python
# now making an empty list to add total number of death in every country
total_deaths_in_country = []

# using for loop to find total death and adding it into the list
for i in countries:
    al = countries_deaths.loc[i,:].sum()
    total_deaths_in_country.append(al)
    
# making dictionary of country name and total number of deaths
country_data = {'Country':countries, 'Deaths':total_deaths_in_country}

# passing it into dataframe
country_df = pd.DataFrame(country_data)

# sorting data in descending order of death counts
sorted_country_df = country_df.sort_values(by = 'Deaths', ascending = False)
```

```python
# let's check top 20 countries with highest number of deaths
plt.figure(figsize = (15,10))
plt.title('Countries with highest death counts')
sns.barplot(x = 'Deaths', y = 'Country', data = sorted_country_df.iloc[:20, :], orient = 'h')
plt.show()
```

Most number of death people are from these top 20 countries.

```python
# let's check 10 countries with least number of deaths
plt.figure(figsize = (15,10))
plt.title('Countries with least death counts')
sns.barplot(x = 'Deaths', y = 'Country', data = sorted_country_df.iloc[-10:, :], orient = 'h')
plt.show()
```

There are least people from these countries

### let's check countries with most and least death due to particular disease

```python
# first let's add all causes in a list
causes = list(countries_deaths.columns)
```

```python
# calling our dataframe which we already have with total death counts due to every cause of every country
df = countries_deaths
```

```python
# using for loop to sort every cause death numbers and plot top 10 countries of death numbers due to every cause
for i in causes:
    sor = countries_deaths.loc[:,i].sort_values(ascending = False)
    d = pd.DataFrame(sor)
    d.reset_index(inplace = True)
    print('Top 10 countries with most number of deaths due to',i)
    sns.barplot(x = i, y = 'Country/Territory', data = d.iloc[:10,:])
    plt.show()
```

Observation: India, China and certain more countries have most number of death due to most of the disease

```python
# using for loop to sort every cause death numbers and plot top 10 countries of death numbers due to every cause
for i in causes:
    sor = countries_deaths.loc[:,i].sort_values()
    d = pd.DataFrame(sor)
    d.reset_index(inplace = True)
    print('Countries with least number of deaths due to',i)
    sns.barplot(x = i, y = 'Country/Territory', data = d.iloc[:10,:])
    plt.show()
```

Observations: There are certain disease due to which there is not any death occurs in certain countries.

### Let's check death by year

```python
# adding years in list
year = list(data.Year.unique())

# making dataframe by year group
year_data = data.groupby('Year').sum()

# making empty list to add total number of death in every year in list
total_death_in_year = []

# using for loop to find total number of deaths in every year and adding them into list
for i in year:
    total = year_data.loc[i,:].sum()
    total_death_in_year.append(total)
```

```python
# making dictionary of year and death 
year_death = {'Year':year, 'No. of Death':total_death_in_year}

# making dataframe of dictionary
df1 = pd.DataFrame(year_death)
```

```python
# plotting data into graph
plt.figure(figsize = (15,10))
sns.lineplot(x = 'Year', y = 'No. of Death', data = df1, marker = 'o')
plt.show()
```

### Let's check most death cause disease in every year

```python
# let's convert rows into columns and vice vers of our dataframe year_data
df2 = year_data.transpose()

# adding different years into a list
year = list(df2.columns)
```

```python
# using for loop to find every year data and plotting them into graph
for i in year:
    sort = df2.loc[ : ,i].sort_values(ascending = False)
    s = pd.DataFrame(sort)
    s.reset_index(inplace = True)
    print('Top 15 death cause in year',i)
    sns.barplot(x = i, y = 'index', data = s.iloc[ :15 ,])
    plt.show()
```

Observation: After analysing every graph we can say that there are certain disease which are the factor of most death in most of the years.

### least death cause disease in every year.

```python
# using for loop to find every year data and plotting them into graph
for i in year:
    sort = df2.loc[ : ,i].sort_values()
    s = pd.DataFrame(sort)
    s.reset_index(inplace = True)
    print('Least death cause disease in year',i)
    sns.barplot(x = i, y = 'index', data = s.iloc[ :15 ,])
    plt.show()
```
